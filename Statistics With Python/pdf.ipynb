# Probability Distribution Function

### Probability Mass Function: (Discrete Random Variable)

```python
import matplotlib.pyplot as plt
import numpy as np

data = [[1, 1],
        [2, 1],
        [3, 1],
        [4, 1],
        [5, 1],
        [6, 1]]
print(data)
x = [d[0] for d in data]
y = [d[1]/sum([d[1] for d in data]) for d in data]

plt.stem(x, y)
plt.xlabel('Die Value')
plt.ylabel('PMF')
plt.title('Probability Mass Function')
plt.show()

```

#### Cumulative Distribution Function: 

```python
cdf = np.cumsum(y)
plt.plot(x, cdf)
plt.xlabel('Die Value')
plt.ylabel('CDF')
plt.title('Cumulative Distribution Function')
plt.show()
```

### Probability Density Function: (continuous Random Variable)

```python
import matplotlib.pyplot as plt

mean = 50
std_dev = 10
sample_size = 1000

data = np.random.normal(mean, std_dev, sample_size)
print(data)

count, bins, ignored = plt.hist(data, 30, density=True)

plt.plot(bins, 1/(std_dev * np.sqrt(2 * np.pi)) * 
         np.exp( - (bins - mean)**2 / (2 * std_dev**2) ),
         linewidth=2, color='r')
plt.xlabel('Value')
plt.ylabel('PDF')
plt.title('Probability Density Function')
plt.show()
```

#### Cumulative Distribution Function: 

```python
plt.plot(bins, 1/(std_dev * np.sqrt(2 * np.pi)) * 
         np.exp( - (bins - mean)**2 / (2 * std_dev**2) ).cumsum() * np.diff(bins)[0], 
         linewidth=2, color='r')

plt.xlabel('Value')
plt.ylabel('CDF')
plt.title('Cumulative Distribution Function')
plt.show()
```

## Types of Distribution:

### Normal Distribution:

```python
import numpy as np 
import seaborn as sns
```

```python
normal_dist_data = np.random.normal(0.5, 0.2, 1000)
normal_dist_data
```

```python
sns.histplot(normal_dist_data, kde= True)
```

### Standard Normal Distribution:

```python
import numpy as np

# create a normal distribution with mean=10 and standard deviation=2
normal_data = np.random.normal(loc=10, scale=2, size=100)
sns.histplot(normal_data, kde = True)
```

```python
# calculate the z-scores using the mean and standard deviation of the normal distribution
z_scores = (normal_data - np.mean(normal_data)) / np.std(normal_data)

# the resulting z-scores will have a mean of 0 and a standard deviation of 1
sns.histplot(z_scores, kde=True)
```

### Log Normal Distribution:

```python
# log normal distribution:
mu = 3.0 # mean 
sigma = 1.0 # standard deviation 

log_normal_dist_data = np.random.lognormal(mu, sigma, 1000)
log_normal_dist_data
```

```python
sns.histplot(log_normal_dist_data, kde=True)
```

```python
# apply transformation:
log_normal_to_normal = np.log(log_normal_dist_data)
```

```python
sns.histplot(log_normal_to_normal, kde=True)
```
